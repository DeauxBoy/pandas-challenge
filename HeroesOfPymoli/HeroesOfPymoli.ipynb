```python
purchase_data_df["Gender"].value_counts().mean()

def change(number):
    return "${:,.2f}".format(number)

male_purchase_count = 652
female_purchase_count = 113
other_purchase_count = 15

male_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Male"]
female_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Female"]
other_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed"]

m_total_price = change(male_df["Price"].sum())
f_total_price = change(female_df["Price"].sum())
o_total_price = change(other_df["Price"].sum())

m_avg_price = change(male_df["Price"].mean())
f_avg_price = change(female_df["Price"].mean())
o_avg_price = change(other_df["Price"].mean())

m_atpp = change(male_df["Price"].sum() / male_df["SN"].nunique())
f_atpp = change(female_df["Price"].sum() / female_df["SN"].nunique())
o_atpp = change(other_df["Price"].sum() / other_df["SN"].nunique())

gender_stats = pd.DataFrame({"":["Female", "Male", "Other / Non Disclosed"],
                                    "Purchase Count": [female_purchase_count, male_purchase_count, other_purchase_count],
                                    "Average Purchase Price": [m_avg_price, f_avg_price, o_avg_price],
                                    "Total Purchase Value": [m_total_price, f_total_price, o_total_price],
                                    "Avg Total Purchase per Person": [f_atpp, m_atpp, o_atpp]})

gender_stats
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
no_dupe = no_dupe.copy()
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age_lables = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
no_dupe["Age Range"]= pd.cut(no_dupe["Age"], bins,labels = age_lables)

all_rows = len(no_dupe)
count_age_range = no_dupe.groupby("Age Range")["SN"].count()
new_df = pd.DataFrame(count_age_range)
new_df["%"] = round(new_df["SN"] / all_rows * 100, 2)
new_df["%"] = new_df["%"].map("{:,.2f}%".format)
new_df.columns = ["Total Count", "Percentage of Players"]
new_df
```
